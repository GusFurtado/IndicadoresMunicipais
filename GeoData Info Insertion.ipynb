```python
# Packages
import json
from DadosAbertosBrasil import ibge
import pandas as pd


# Baixar GeoData
with open('assets/geojs-42-mun.json', 'r', encoding='utf-8') as j:
    geo = json.load(j)
    
    
# Definir lista de parâmetros do SIDRA
KPIS = {
    
    'Rede Geral de Abastecimento de Água': {
        'agregado': 3218,
        'periodo': 2010,
        'variaveis': 1000096,
        'classificacoes': {61: 92853}
    },
    
    'Sem Banheiro ou Sanitário': {
        'agregado': 3218,
        'periodo': 2010,
        'variaveis': 1000096,
        'classificacoes': {299: 10006}
    },
    
    'Lixo Coletado': {
        'agregado': 3218,
        'periodo': 2010,
        'variaveis': 1000096,
        'classificacoes': {67: 2520}
    },
    
    'Energia Elétrica': {
        'agregado': 3218,
        'periodo': 2010,
        'variaveis': 1000096,
        'classificacoes': {309: 3011}
    }
    
}


# Atualizar GeoData com dados do SIDRA
for i in KPIS:
    
    print(i)
    kpi = KPIS[i]

    sidra = ibge.Sidra(agregado = kpi['agregado'],
                       periodos = kpi['periodo'],
                       variaveis = kpi['variaveis'],
                       classificacoes = kpi['classificacoes'],
                       localidades = {'N6': 'all'})

    print('- Baixando dados do SIDRA')
    resultados = sidra.rodar()

    print('- Inserindo dados no JSON')
    for id_geo in geo['features']:
        for id_res in resultados[0]['resultados'][0]['series']:
            if id_geo['properties']['id'] == id_res['localidade']['id']:
                try:
                    value = float(id_res['serie'][str(kpi['periodo'])])
                except:
                    value = 0
                id_geo['properties'].update({i: value})
                break
    
    print()
    
    
# Salvar novo JSON
with open('assets/sc_geo.json', 'w') as fp:
    json.dump(geo, fp)
```
