```python
import requests

from bs4 import BeautifulSoup
import pandas as pd

from aux_functions import lista_municipios
```

```python
MUNS = lista_municipios()
```

```python
mun = 'Joinville'

link = MUNS.loc[MUNS[MUNS.Município == mun].index[0], 'href']
s = requests.Session()
html = s.get(r'https://pt.wikipedia.org' + link).text
soup = BeautifulSoup(html, 'html.parser')
table = soup.find('table', 'infobox_v2')

wiki_scrap = [[td.text for td in tr.find_all('td')] for tr in table.find_all('tr')]
pd.DataFrame(wiki_scrap)
```
